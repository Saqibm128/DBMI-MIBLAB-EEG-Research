```python
import keras
from keras.models import Model, Sequential
from keras.layers import Input, LSTM, Dense, Activation, Conv2D, Concatenate, Dropout, MaxPool2D, Conv3D, Flatten, LeakyReLU, BatchNormalization
```

```python
x_in = np.random.rand(100, 10)
y1 = np.random.rand(100) + 5 * x_in[:, 5] + 7 * x_in[:, 2] ** 2
y1 = y1.reshape(100, 1)
y2 = x_in[:, 4] ** 3 + x_in[:, 7] ** 0.5 + np.random.rand(100)
y2 = y2.reshape(100, 1)
y3 = keras.utils.to_categorical(np.round(x_in[:,9]), num_classes=2)
```

```python
y2 = (y2 - y2.min()) / (y2.max() - y2.min())
```

```python
x = Input(shape=(10,))
y = Dense(30)(x)
y = Dense(30)(y)
y = Dense(30)(y)
y = Dense(30)(y)
y_pred_1 = Dense(1, activation="linear")(y)
y_pred_2 = Dense(1, activation="sigmoid")(y)
y_pred_3 = Dense(2, activation="sigmoid")(y)
```

```python
model = Model(inputs=x, outputs=[y_pred_1, y_pred_2, y_pred_3])
```

```python
model.compile("adam", loss=["mean_squared_error", "mean_squared_error", "categorical_crossentropy"], loss_weights=[0.3, 0.3, 0.3])
```
